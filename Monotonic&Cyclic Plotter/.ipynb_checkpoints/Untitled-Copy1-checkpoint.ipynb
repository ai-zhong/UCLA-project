```python
%matplotlib inline
%config InlineBackend.figure_format='retina'
import matplotlib.pyplot as plt
import matplotlib.backends.backend_pdf

import numpy as np
import pandas as pd
from ipywidgets import *
from IPython.display import display
from IPython.display import clear_output
from matplotlib import rc
from scipy.signal import find_peaks_cwt
import os

font = {'color':  'black',
        'weight': 'normal',
        'size': 12}

dir_root = os.getcwd()
dir_root_cyclic = dir_root + "/Cyclic/"
dir_root_monotonic = dir_root + "/Monotonic/"
test_dir = dir_root_cyclic

path = test_dir + 'SWnew/Shear.txt'
data = np.genfromtxt(path, delimiter=',')
driving = data[2:,22]
start_index = np.where(driving>0)[0][0]

time = data[start_index: ,0]-data[start_index,0]
driving = data[start_index:,22]
shear_stress = data[start_index: ,17]
shear_strain = data[start_index: ,14]
vertical_stress = data[start_index:,18]
initial_vertical_stress = vertical_stress[0]

for n in range(1,len(driving)):
    if (driving[n]>driving[n-1]) & (driving[n]>driving[n+1]):
        first_peak=n
        peak1=shear_stress[first_peak+1]
        peak2=shear_stress[(first_peak+1)*3]
        period = time[n-1]*4
        break

print period
peak_avg =(abs(peak1)+abs(peak2))/2
cycle_all = time/period

plt.plot(cycle_all, shear_strain)
# plt.plot(cycle_all, driving, '--', color = '0.75')
plt.ylabel('Shear strain')
```

```python
lab_csr = 'SWnew; strain:3%'
f, axes = plt.subplots(1,2,figsize=(10,5))
ss_for_csr = 3

for n in range(0,len(shear_strain)):
    if int(shear_strain[n]) == int(ss_for_csr):
        index_min = n
        break
    
time_def_percent = time[index_min]

CSR = round(float(peak_avg/initial_vertical_stress),3)
Number_of_Cycle = round(float(time_def_percent/period),2)    

axes[0].semilogx(Number_of_Cycle, CSR, 'o', label = lab_csr)
axes[0].set_xlabel('Number of Cycle, N (logscale)', fontdict=font)
axes[0].set_ylabel('Cyclic Stress Ratio, CSR', fontdict=font)
axes[0].set_title('CSR Plot', fontdict=font)
axes[0].grid(True, which='both')
axes[0].tick_params(axis='x', which='minor')
axes[0].legend(bbox_to_anchor=(1, 1),prop={'size':8}) 

plt.plot(cycle_all[:index_min], shear_strain[:index_min])

print (lab_csr + '; CSR = '+str(CSR)+'; Number of Cycle: '+str(Number_of_Cycle))
```
